```python
# Looping and Functons 
# q1 . Compose a pytohn FUnction which returns cubes of Natural Numbers

def cub (num):
    cobes = []
    for i in range(1,num+1):
        value = i**3
        cobes.append(value)
    return cobes
---------------
num = input("ENter a number: ")
cobes = cub(int(num))
print(cobes)
```
