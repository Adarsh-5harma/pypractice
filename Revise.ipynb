```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data: Payload weight (kg) vs Fuel used (liters)
X = np.array([100, 200, 300, 400, 500])  # payload
Y = np.array([500, 700, 900, 1100, 1300])  # fuel

# Step 1: Calculate the coefficients for y = b0 + b1*x
n = len(X)
b1 = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2) - (np.sum(X))**2) 
b0 = (np.sum(Y) - b1*np.sum(X)) / n

print(f"Linear Model: Fuel = {b0:.2f} + {b1:.2f} * Payload")

# Step 2: Predict fuel for a new payload
new_payload = 600
predicted_fuel = b0 + b1 * new_payload
print(f"Predicted fuel for {new_payload} kg payload: {predicted_fuel:.2f} liters")

# Step 3: Plot
plt.scatter(X, Y, color='blue', label='Data')
plt.plot(X, b0 + b1*X, color='red', label='Regression Line')
plt.xlabel("Payload (kg)")
plt.ylabel("Fuel (liters)")
plt.title("Linear Regression from Scratch")
plt.legend()
plt.show()
```
